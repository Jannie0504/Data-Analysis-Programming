# 4. 데이터 탐색

### 데이터 재구조화

```python
# 연속형 나이 변수 범주화
bins = [-1, 19, 39, 59, 79, 99]
df_Age = pd.cut(df['Age'], bins, labels = ['child', 'young', 'middle', 'prime', 'old'])
df['Age_class'] = df_Age
df
# child = 0 ~ 19세
# young = 20 ~ 39세
# middle = 40 ~ 59세
# prime = 60 ~ 79세
# old = 80 ~ 99세
```

```python
# 연속형 연간 소득 변수 범주화
bins = [-1, 50000, 100000, 150000, max(df['Annual Income ($)'])]
df_Annual_Income = pd.cut(df['Annual Income ($)'], bins, labels = ['bronze', 'sliver', 'gold', 'diamond'])
df['Income_class'] = df_Annual_Income
df
# bronze = 0 ~ 50,000$
# sliver = 50,001 ~ 100,000$
# gold = 100,001 ~ 150,000$
# diamond = 150,001 ~
```

```python
# 연속형 지출 점수 변수 범주화
bins = [0, 25, 50, 75, 100]
df_score = pd.cut(df['Spending Score (1-100)'], bins, labels = ['D', 'C', 'B', 'A'])
df['Spending_class'] = df_score
df
# D = 0 ~ 25점
# C = 26 ~ 50점
# B = 51 ~ 75점
# A = 76점 ~ 100점
```

```python
# 연속형 업무 경험 변수 범주화
bins = [-1, 0, 3, 8, 13, 17]
df_Work_Experience = pd.cut(df['Work Experience'], bins, labels = ['no', 'newcomer', 'junior', 'senior', 'leader'])
df['Work_class'] = df_Work_Experience
df
# no = 경력 없음
# newcomer = 1 ~ 3년차
# junior = 4 ~ 8년차
# senior = 9 ~ 13년차
# leader = 14 ~ 17년차
```

```python
# 연속형 업무 경험 변수 범주화
bins = [0, 2, 5, 9]
df_Family_Size = pd.cut(df['Family Size'], bins, labels = ['small', 'medium', 'large'])
df['Family_class'] = df_Family_Size
df
# small = 1 ~ 2명
# medium = 3 ~ 5명
# large = 6 ~ 9명
```

##### D : 여성 > 남성 | young         | silver       | junior           | medium

### 범주형 * 범주형

#### 각 column 그룹별 지출 점수와의 연관성

```python
# 지출 점수별 성별 분포 확인
sns.catplot(x='Gender', kind='count', hue='Spending_class', data=df)
```

##### 여성이 남성보다 인원수가 많음
##### 남성은 D, C, B 그룹의 인원수가 점점 증가
##### 여성은 D, B, A 그룹의 인원수가 점점 증가, C 그룹의 인원수가 압도적으로 많음

```python
def annotateBars(*args, **kwargs):
    ax = plt.gca()
    for p in ax.patches:
        ax.annotate(format(p.get_height()),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center',
                    xytext = (0, 10),
                    textcoords = 'offset points')
```

```python
# 지출 점수별 나이 분포 확인
g = sns.catplot(x='Spending_class', kind='count', col='Age_class', data=df)
g.map(annotateBars)
plt.show()
```

##### 어린 나이대(child, young)를 제외하고 나머지 나이대에선 C 그룹의 인원수가 가장 많음

```python
# 지출 점수별 연간 소득 분포 확인
g = sns.catplot(x='Spending_class', kind='count', col='Income_class', data=df)
g.map(annotateBars)
plt.show()
```

##### bronze 그룹을 제외한 나머지 그룹에서는 모두 C 그룹의 인원수가 가장 많음

```python
# 지출 점수별 업무 경험 분포 확인
g = sns.catplot(x='Spending_class', kind='count', col='Work_class', data=df)
g.map(annotateBars)
plt.show()
```

##### 경력이 아예 없는 no 그룹을 제외하고 나머지 그룹에서는 모두 C 그룹의 인원수가 가장 많음
##### 경력이 아예 없는 no 그룹은 지출 점수 그룹별 인원수가 증가함

```python
# 지출 점수별 가족 규모 분포 확인
g = sns.catplot(x='Spending_class', kind='count', col='Family_class', data=df)
g.map(annotateBars)
plt.show()
```

##### 모든 그룹에서 C 그룹의 인원수가 가장 많음 (medium 그룹은 C와 B 그룹이 동일)
##### 그 뒤로 B 그룹의 인원수가 많음
